## Q8 Check the number of rentals for each customer for May

### OPTION 1. Establishing the connection with MySQL -- EASIER OPTION

```python
connection_string = 'mysql+pymysql://root:' + password + '@localhost/sakila'
engine = create_engine(connection_string)
datamay = pd.read_sql_query('SELECT *, count(rental_id) AS N_rental_May FROM rental_may \
GROUP BY customer_id \
ORDER BY N_rental_May DESC;', engine)
datamay.head()
```

## Q9 Check the number of rentals for each customer for June 
### OPTION 3. Getting a Panda's dataframe (df) called "data" out of a MySQL query -- good for long queries

```python
engine.execute("USE sakila")
query = 'SELECT *, count(rental_id) AS N_rental_June FROM rental_june GROUP BY customer_id ORDER BY N_rental_June DESC;'
datajune = pd.read_sql_query(query, engine)
datajune.head()
```

## Q10 Write a function that checks if customer borrowed more or less books in the month of June as compared to May.

```python
import pandas as pd
import numpy as np
```

```python
merged = datamay.merge(datajune, how='inner', on='customer_id')
merged
```

```python
may = 0
june = 0
for value in range(len(merged['customer_id'])):
    if merged.N_rental_May[value] > merged.N_rental_June[value]:
        may += 1
    else:
        june +=1
print ('On May,',may,'customers made more rentals');
print ('On June,',june,'customers made more rentals');
if may > june:
    print( 'May wins!')
else:
    print('June wins!')
```
